```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# Use the following data for this assignment:
import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

```python
mean = df.mean(axis = 1)
std = df.std(axis = 1)

n= df.shape[1]
yerr = std / np.sqrt(n) * stats.t.ppf(1-0.05/2, n - 1)

plt.figure()
plt.bar(range(df.shape[0]), mean, yerr = yerr,align='center', alpha=0.7, ecolor='black', capsize=10,color=['red', 'green', 'blue', 'cyan']) 
plt.xlabel('year')
#plt.style.use('fivethirtyeight')
plt.ylabel('Y-label')
plt.title('Custom Visualization')
my_xticks = ['1992', '1993', '1994', '1995']
plt.xticks((0, 1, 2, 3), my_xticks)
plt.grid(True)
plt.tight_layout()
plt.savefig('Customvisu.png')
plt.show() 
```
